```python
import GenomeUtils
from pathlib import Path
```

```python
downloader = GenomeUtils.Downloaders.EnsemblGenomeDownloader(
    assembly_id="GRCh38",
    ensembl_release=115,
    species="homo_sapiens",
    genomes_root_dir=Path("./data/genomes"),
)

paths = downloader.download()
```

```python
human_genome, scaffold_genome = (
    GenomeUtils.Genome.GenomeBuilder(id="GRCh38", species="Homo sapiens", name="Human")
    .with_dna_fasta(paths["dna"])
    .with_cdna_fasta(paths["cdna"])
    .with_gtf_file(paths["annotation"])
    .build()
)
```
